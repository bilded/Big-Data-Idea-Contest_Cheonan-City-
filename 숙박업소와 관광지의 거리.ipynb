```python
from geopy.geocoders import Nominatim
import pandas as pd

# 주소 정리 함수
def remove_dash_zero(address):
    a = address.split(' ')
    if a[-1].endswith('-0'):
        a[-1] = a[-1].replace('-0', '')
    return " ".join(a)

# 주소 목록
addresses = [
    "충남 천안시 동남구 만남의길 19-10",
    "충남 천안시 동남구 해사동길 15-0",
    "충남 천안시 동남구 신부2길 31-0",
    "충남 천안시 동남구 광풍로 703-38",
    "충남 천안시 동남구 덕성1길 147-0"
]

# 주소 정리
cleaned_addresses = [remove_dash_zero(addr) for addr in addresses]

# Nominatim 객체 생성
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        if geo is None:
            print(f"주소를 찾을 수 없습니다: {address}")
            return [0, 0]
        x_y = [geo.latitude, geo.longitude]
        return x_y
    except Exception as e:
        print(f"오류 발생: {e}")
        return [0, 0]

# 위도와 경도 리스트 초기화
latitude = []
longitude = []

# 각 주소에 대해 위도와 경도 계산
for addr in cleaned_addresses:
    lat, lon = geocoding(addr)
    latitude.append(lat)
    longitude.append(lon)

# DataFrame 생성
address_df = pd.DataFrame({
    '주소': cleaned_addresses,
    '위도': latitude,
    '경도': longitude
})

# DataFrame을 CSV로 저장
address_df.to_csv('cleaned_address_with_coordinates.csv', index=False)

# 결과 출력
print(address_df)
```
